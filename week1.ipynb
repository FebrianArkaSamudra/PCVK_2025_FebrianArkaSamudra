Step 1

```python
import numpy as np
import pandas as pd
import cv2 as cv
from google.colab.patches import cv2_imshow  # for image display
from skimage import io
from skimage import transform
from PIL import Image
import matplotlib.pylab as plt
```

Step 2

```python
# Membuat list untuk menyimpan url dari beberapa image
urls = [
    "https://i.pinimg.com/1200x/41/dc/bc/41dcbce9aa4e5210c6c00774e1b10cb4.jpg",
    "https://i.pinimg.com/736x/3d/84/ea/3d84eaf18f7aeee80572413ba1a5c6ba.jpg",
    "https://i.pinimg.com/736x/43/5a/26/435a26f096e384d5cf8e97ca6096efd6.jpg"
]
# baca dan tampilkan image
# Loop pada tiap url image,beberapa image dapat disimpan pada list
for url in urls:
        image = io.imread(url)  #read image
        image = cv.resize(image, (0, 0), fx=0.5, fy=0.5)  # resize image to half size
        image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)   # convert color to RGB
        final_frame = cv.hconcat([image, image_2])       # Concatenate image
        cv2_imshow(final_frame)                          # Show image
        print('\n')
```

Step 3

```python
tinggi = image_2.shape[0]
lebar = image_2.shape[1]
print("resolusi image: tinggi x lebar = ",tinggi, " x ",lebar)
cv2_imshow(image_2)
```

Step 4

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

for y in range(lebar):
    image_3[int(tinggi / 2), y] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

min_side = min(tinggi, lebar)
for x in range(min_side):
    image_3[x, x] = [255, 255, 255]

for x in range(min_side):
    image_3[x, lebar - 1 - x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
y_center = int(tinggi/2)
line_length = 100
x_start = int((lebar - line_length)/2)
x_end = x_start + line_length
for x in range(x_start, x_end):
    image_3[y_center, x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```

```python
image_2 = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_3 = cv.cvtColor(image, cv.COLOR_BGR2RGB)

box_size = 100
x_center = lebar // 2
y_center = tinggi // 2
x1 = x_center - box_size // 2
y1 = y_center - box_size // 2
x2 = x_center + box_size // 2
y2 = y_center + box_size // 2

# kotak
for y in range(y1, y2):
    for x in range(x1, x2):
        image_3[y, x] = [255, 255, 255]

final_frame = cv.hconcat([image_2, image_3])
cv2_imshow(final_frame)
```
